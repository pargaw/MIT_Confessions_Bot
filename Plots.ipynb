```python
from matplotlib import pyplot as plt 
'''
Perplexity AND Loss for ALL models (12 plots!!)
    Baselines (LSTM vs. GRU)  (char-char)
    Baselines (LSTM vs. GRU) (word-word)
    Char-char vs word-word (baseline LSTM)
    Char-Char vs. word-word (baseline GRU)
    Char-char vs word-word (conv LSTM)
    Reactions: likes-only and all-reactions (best model -LSTM)
'''

# load all logs
lstm_char= 'lstm_512_charchar_baseline.log'
gru_char = 'gru_512_charchar_baseline.log'
lstm_word= 'lstm_512_wordword.log'
gru_word = 'gru_512_wordword.log'
c_lstm_char= 'conv_lstm_512_charchar.log'
c_lstm_word = 'c_lstm_512_wordword.log'
c_lstm_char_reacts = 'c_lstm_512_charchar_react.log'
c_lstm_char_likes = 'lstm_charchar_likes.log'
lstm_char_reacts = 'lstm_512_charchar_reacts.log'
```

```python
# Baselines (LSTM char vs. GRU char)  
epoch_X = {}
train_perplex = {}
val_perplex = {}
train_loss = {}
val_loss = {}

fig = plt.figure()
ax = plt.axes()

for name in [lstm_char,gru_char]:
    f = open(name,'r')
    for line in f.readlines()[1:]:
        data = line.strip().split(",")
        if name in epoch_X:
            epoch_X[name].append(int(data[0]))
        else:
            epoch_X[name] = [int(data[0])]
        
        if name in train_loss and name in val_loss:
            train_loss[name].append(float(data[1]))
            val_loss[name].append(float(data[3]))
        else:
            train_loss[name] = [float(data[1])]
            val_loss[name] = [float(data[3])]
            
        if name in train_perplex and name in val_perplex:
            train_perplex[name].append(2**(float(data[1])))
            val_perplex[name].append(2**(float(data[3])))
        else:
            train_perplex[name] = [2**(float(data[1]))]
            val_perplex[name] = [2**(float(data[3]))]
            
    plt.figure(1)
    if name == lstm_char:
        plt.plot(epoch_X[name], train_perplex[name], color='blue')
        plt.plot(epoch_X[name], val_perplex[name], linestyle='--', color='blue')
    else:
        plt.plot(epoch_X[name], train_perplex[name], color='red')
        plt.plot(epoch_X[name], val_perplex[name], linestyle='--', color='red')
    plt.title('LSTM char vs. GRU char perplexities')

    plt.figure(2)
    if name == lstm_char:
        plt.plot(epoch_X[name], train_loss[name], color='blue')
        plt.plot(epoch_X[name], val_loss[name], linestyle='--', color='blue')
    else:
        plt.plot(epoch_X[name], train_loss[name], color='red')
        plt.plot(epoch_X[name], val_loss[name], linestyle='--', color='red')
    plt.title('LSTM char vs. GRU char losses')
    
    
plt.figure(1)
plt.legend(['LSTM train perplexity', 'LSTM val perplexity','GRU train perplexity', 'GRU val perplexity'], fontsize=8, loc='upper right')
plt.savefig('LSTM_GRU_char_perplexity.jpg')
plt.figure(2)
plt.legend(['LSTM train loss', 'LSTM val loss','GRU train loss', 'GRU val loss'], fontsize=8, loc='upper right')
plt.savefig('LSTM_GRU_char_loss.jpg')
plt.show()
plt.clf()
```
